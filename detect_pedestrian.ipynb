```
import cv2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```
pedestrian_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_fullbody.xml')
```

```
pedestrian_cascade
```

```
picture = cv2.imread("ped_detect.jpeg")
picture
picture.shape
```

```
picture = imutils.resize(picture, width=500)
picture.shape
```

```
plt.imshow(picture)
plt.show()
```

```
picture = cv2.cvtColor(picture, cv2.COLOR_BGR2RGB)
```

```
gray1 = cv2.cvtColor(picture, cv2.COLOR_BGR2GRAY)
```

```
pedestrian = pedestrian_cascade.detectMultiScale(gray1, 1.1, 5, minSize=(40,50))
```

```
for (x,y,w,h) in pedestrian:
        cv2.rectangle(pic, (x, y), (x+w, y+h), (255, 0, 0), 2)
```
